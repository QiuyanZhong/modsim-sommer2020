Beispiele für Approximationen für $n_{\infty}$ und $\tau_n$ die folgende Funktionen:
$$n_\infty (V) =\frac{1}{1+\exp \left( \frac{V_0 - V}{k} \right) }$$ 
mit den freien Parametern $V_0$, $k$,

```octave
function value=ninfty_model(v, V0, k)
    value = 1.0./(1.0+exp((V0.-v)/k));
endfunction
```

```octave
Vfine = linspace(-10,10,500);
plot(Vfine, ninfty_model(Vfine, -5, 1))
```

$$ \tau_n(V) \approx \tau_0 + (\tau_1 - \tau_0) \exp \left(-\frac{\|V-V_{max}\|^2}{\sigma^2} \right) $$
mit den Parametern $\tau_0$,  $\tau_1$, $V_{max}$ und $\sigma^2$.

```octave
function value=tau_model(v, Vmax, sigma2, tau1, tau0)
    value = tau0 + (tau1-tau0)*exp(-(v.-Vmax).*(v.-Vmax) ./ sigma2) ;
endfunction
```

```octave
Vfine = linspace(-10,10,500);
plot(Vfine, tau_model(Vfine, 5, 2, 4, -1))
```
